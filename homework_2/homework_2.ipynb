```python
from matplotlib import pyplot as plt
import numpy as np
```

## 8.4.3

```python
x = np.linspace(0.000000001,.999999999)
g_x = x*(1-x)+(1-x)*x
d_x = -(x*np.log(x) + (1-x)*np.log(1-x))
class_max = []
for i in x:
    class_max.append(1 - max([i,1-i]))
```

```python
plt.plot(x,g_x,'o',label = "Gini Index")
plt.plot(x,d_x,'o',label = "Cross-entropy")
plt.plot(x,class_max,'o',label = "Classification error rate")
plt.xlabel(r'$\hat{p}_{m1}$')
plt.legend()
plt.savefig('plots/8_4_3.png')
```

```python
x1 =np.linspace(-3,1,5000)  # Adjust the range as needed
# Calculate corresponding y values based on the equation
# x = np.linspace(-3,1,5000)
# y = np.linspace(-3,1,5000)

# X, Y = np.meshgrid(x, y)

x2_positive = np.sqrt(4 - (x1 + 1)**2) + 2
x2_negative = -np.sqrt(4 - (x1 + 1)**2) + 2 

# condition = (1 + X)**2 + (2 - Y)**2 > 4
# condition_2 = (1 + X)**2 + (2 - Y)**2 <= 4

# Plot the circle
plt.figure(figsize=(6, 6))  # Optional: set the figure size
plt.plot(x1, x2_positive, color = "red")
plt.plot(x1, x2_negative, color = "red")  # Plot the other half

plt.fill_between(x1,x2_positive,2, color = 'lightcoral', label = "Red Class")
plt.fill_between(x1,x2_negative,2, color = 'lightcoral')

plt.fill_between(x1,x2_positive,8, color = "skyblue")
plt.fill_between(x1,x2_negative,-1, color = "skyblue")

x1b = np.linspace(-4,-3,100)
plt.fill_between(x1b, -1,8, color = "skyblue")
x1c = np.linspace(1,3,100)
plt.fill_between(x1c, -1,8, color =  "skyblue", label = "Blue Class")


for z in zip([0,-1,2,3],[0,1,2,8],["blue","red","blue","blue"]):

    plt.scatter(z[0],z[1], color=z[2], marker='o')
    plt.text(z[0]-.15, z[1]-.15, f'({z[0]}, {z[1]})', fontsize=12, ha='right', va = 'top')


# plt.axhline(y=0, color='black')
# Optional: Add labels and title
plt.xlabel(r'$x_1$')
plt.ylabel(r'$x_2$')

# Equal aspect ratio to ensure a circular shape
plt.gca().set_aspect('equal')

# Display the plot
# plt.savefig('plots/6_18.png')
plt.legend()
plt.savefig('plots/9_7_2.png')
plt.show()
```
